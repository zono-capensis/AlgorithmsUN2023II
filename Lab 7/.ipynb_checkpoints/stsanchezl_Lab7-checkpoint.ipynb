```python
import random

def generate_random_preferences(n):
    men_preferences = []
    women_preferences = []

    for _ in range(n):
        # Genera listas aleatorias de preferencias para hombres y mujeres
        men_prefs = list(range(n))
        random.shuffle(men_prefs)
        men_preferences.append(men_prefs)

        women_prefs = list(range(n))
        random.shuffle(women_prefs)
        women_preferences.append(women_prefs)

    return men_preferences, women_preferences
```

```python
def gale_shapley(men_list, women_list, men_pref, women_pref):
    # Initialization
    n = len(men_list)
    free_men = list(men_list)
    women_partner = [None for i in range(n)]
    next_proposal = [0 for i in range(n)]

    # While there are free men
    while free_men:
        for man in free_men:
            # Propose to the next woman on the list
            woman = men_pref[man][next_proposal[man]]
            next_proposal[man] += 1

            # If the woman is free, engage
            if women_partner[woman] is None:
                women_partner[woman] = man
                free_men.remove(man)
            else:
                # If the woman prefers this man, engage and free the old partner
                if women_pref[woman].index(man) < women_pref[woman].index(women_partner[woman]):
                    free_men.append(women_partner[woman])
                    free_men.remove(man)
                    women_partner[woman] = man

    return women_partner

# Best case: Each person's list of preferences is in the same order
men_pref_best = [[0, 1, 2, 3] for _ in range(4)]
women_pref_best = [[0, 1, 2, 3] for _ in range(4)]
men_list = [0, 1, 2, 3]
women_list = [0, 1, 2, 3]

print(gale_shapley(men_list, women_list, men_pref_best, women_pref_best))

# Worst case: Each man's list of preferences is in the same order, and each woman's list of preferences is in reverse order
men_pref_worst = [[0, 1, 2, 3] for _ in range(4)]
women_pref_worst = [[3, 2, 1, 0] for _ in range(4)]

print(gale_shapley(men_list, women_list, men_pref_worst, women_pref_worst))
```

```python
import random
import matplotlib.pyplot as plt

def gale_shapley(n, men_pref, women_pref):
    men_list = list(range(n))
    women_list = list(range(n))
    free_men = list(men_list)
    women_partner = [None for i in range(n)]
    next_proposal = [0 for i in range(n)]
    iterations = 0

    while free_men:
        for man in free_men:
            woman = men_pref[man][next_proposal[man]]
            next_proposal[man] += 1
            iterations += 1

            if women_partner[woman] is None:
                women_partner[woman] = man
                free_men.remove(man)
            else:
                if women_pref[woman].index(man) < women_pref[woman].index(women_partner[woman]):
                    free_men.append(women_partner[woman])
                    free_men.remove(man)
                    women_partner[woman] = man

    return iterations

def run_experiment(n):
    iterations = []
    for _ in range(100):
        men_pref = [random.sample(range(n), n) for _ in range(n)]
        women_pref = [random.sample(range(n), n) for _ in range(n)]
        iterations.append(gale_shapley(n, men_pref, women_pref))

    print(f"For n={n}: Min={min(iterations)}, Max={max(iterations)}, Avg={sum(iterations)/len(iterations)}")
    plt.hist(iterations, bins=range(min(iterations), max(iterations) + 1, 1), alpha=0.7, label=f'n={n}')

run_experiment(4)
run_experiment(40)

plt.xlabel('Iterations')
plt.ylabel('Frequency')
plt.title('Distribution of Iterations for Gale-Shapley Algorithm')
plt.legend(loc='upper right')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Gale-Shapley Algorithm

def stable_marriage(n, men_preferences, women_preferences):
    # Initialize matches and availability
    matches = {}
    men_available = {man: True for man in range(n)}
    iterations = 0

    # Continue until all men are matched
    while any(men_available.values()):
        iterations += 1
        for man in range(n):
            if men_available[man] and men_preferences[man]:
                # Get the woman according to the man's preference
                woman = men_preferences[man].pop(0)


                # Check if the woman is already engaged
                current_man = matches.get(woman)

                if current_man is None:
                    # Woman is not engaged, match them
                    matches[woman] = man
                    men_available[man] = False
                else:
                    # Woman is engaged, compare preferences
                    current_woman_pref = women_preferences[woman]
                    if current_woman_pref.index(current_man) > current_woman_pref.index(man):
                        # Current man is less preferred, update the match
                        matches[woman] = man
                        men_available[man] = False
                        men_available[current_man] = True

    return matches, iterations


def simulate_and_plot(n, num_frames):
    men_preferences = generate_random_preferences(n)
    women_preferences = generate_random_preferences(n)

    # Store intermediate results for animation
    results = []

    for _ in range(num_frames):
        # Run Gale-Shapley algorithm
        _, iterations = stable_marriage(n, men_preferences, women_preferences)
        results.append((men_preferences.copy(), women_preferences.copy(), iterations))


    return results

def update(frame, men_preferences, women_preferences, sc, text):
    # Update the scatter plot and text for each frame
    men_text = [f'Man {i}\n{men_preferences[i]}' for i in range(len(men_preferences))]
    women_text = [f'Woman {i}\n{women_preferences[i]}' for i in range(len(women_preferences))]

    text.set_text(f'Iteration: {frame + 1}')
    sc.set_offsets([(0, i) for i in range(len(men_preferences))])
    text.set_text(f'Iteration: {frame + 1}')
    text.set_text(f'Iteration: {frame + 1}')
    text.set_text(f'Iteration: {frame + 1}')



# Simulate and plot the animation
n = 4
num_frames = 10

results = simulate_and_plot(n, num_frames)

# Set up the initial plot
fig, ax = plt.subplots()
sc = ax.scatter([0] * n, range(n), s=200, c='blue', marker='o')
text = ax.text(0.5, 1.05, '', transform=ax.transAxes, fontsize=10, ha='center')

# Create the animation
ani = FuncAnimation(fig, update, frames=num_frames, fargs=results, blit=False)

# Save the animation as a GIF
ani.save('gale_shapley_animation.gif', writer='imagemagick', fps=1)

# Display the plot
plt.show()
```
